```python
import pandas as pd
import matplotlib.pyplot as plt

def generate_car_matrix(dataset):
   
    car_matrix = pd.pivot_table(dataset, values='car', index='id_1', columns='id_2', fill_value=0)
    
    # Set diagonal values to 0
    car_matrix.values[[range(len(car_matrix))]*2] = 0
    
    return car_matrix

# Adjust the file path accordingly based on your directory structure
dataset = pd.read_csv(r"C:\Users\amrit\Downloads\MapUp-Data-Assessment-F-main\MapUp-Data-Assessment-F-main\datasets\dataset-1.csv")
result_matrix = generate_car_matrix(dataset)

# Print the result DataFrame
df = pd.DataFrame(result_matrix)
df
```

# Question 2: Car Type Count Calculation

```python
import pandas as pd

def get_type_count(dataset):
    # Create a new categorical column 'car_type' based on specified conditions
    dataset['car_type'] = pd.cut(dataset['car'],
                                 bins=[float('-inf'), 15, 25, float('inf')],
                                 labels=['Low', 'Medium', 'High'],
                                 right=False)
    
    # Calculate the count of occurrences for each 'car_type' category
    type_count = dataset['car_type'].value_counts().to_dict()
    
    # Sort the dictionary alphabetically based on keys
    type_count = dict(sorted(type_count.items()))
    
    return type_count

# Adjust the file path accordingly based on your directory structure
dataset = pd.read_csv(r"C:\Users\amrit\Downloads\MapUp-Data-Assessment-F-main\MapUp-Data-Assessment-F-main\datasets\dataset-1.csv")
result_count = get_type_count(dataset)

# Print the result dictionary
print(result_count)
```

# Question 3: Bus Count Index Retrieval

```python
import pandas as pd

def get_bus_indexes(dataset):
    # Calculate the mean value of the 'bus' column
    bus_mean = dataset['bus'].mean()
    
    # Identify indices where 'bus' values are greater than twice the mean value
    bus_indexes = dataset[dataset['bus'] > 2 * bus_mean].index.tolist()
    
    bus_indexes.sort()
    
    return bus_indexes

# Adjust the file path accordingly based on your directory structure
dataset = pd.read_csv(r"C:\Users\amrit\Downloads\MapUp-Data-Assessment-F-main\MapUp-Data-Assessment-F-main\datasets\dataset-1.csv")
result_indexes = get_bus_indexes(dataset)

# Print the result list of indices
print(result_indexes)
```

# Question 4: Route Filtering

```python
import pandas as pd

def filter_routes(dataset):
    # Calculate the average value of the 'truck' column for each route
    route_avg_truck = dataset.groupby('route')['truck'].mean()
    
    # Filter routes where the average value of 'truck' column is greater than 7
    selected_routes = route_avg_truck[route_avg_truck > 7].index.tolist()
    
    selected_routes.sort()
    
    return selected_routes

# Adjust the file path accordingly based on your directory structure
dataset = pd.read_csv(r"C:\Users\amrit\Downloads\MapUp-Data-Assessment-F-main\MapUp-Data-Assessment-F-main\datasets\dataset-1.csv")
result_routes = filter_routes(dataset)

# Print the result list of selected routes
print(result_routes)
```

# Question 5: Matrix Value Modification

```python
import pandas as pd

def multiply_matrix(result_df):
    # Create a copy of the input DataFrame to avoid modifying the original
    modified_df = result_df.copy()
    
    # Apply the specified logic to modify values in the DataFrame
    modified_df = modified_df.applymap(lambda x: x * 0.75 if x > 20 else x * 1.25)
    
    # Round values to 1 decimal place
    modified_df = modified_df.round(1)
    
    return modified_df

# Adjust the file path accordingly based on your directory structure
result_df = pd.read_csv(r"C:\Users\amrit\Downloads\MapUp-Data-Assessment-F-main\MapUp-Data-Assessment-F-main\datasets\dataset-1.csv")  # Replace with the correct file path
modified_result_df = multiply_matrix(result_df)

# Print the modified DataFrame
print(modified_result_df)
```

# Question 6: Time Check

```python
import pandas as pd

def time_check(df):
    # Convert timestamp columns to datetime objects
    df['startTime'] = pd.to_datetime(df['startTime'])
    df['endTime'] = pd.to_datetime(df['endTime'])

    # Calculate time differences for each row
    df['time_diff'] = (df['endTime'] - df['startTime']).dt.total_seconds()

    # Group by (`id`, `id_2`) pairs
    grouped_data = df.groupby(['id', 'id_2'])

    # Check if timestamps cover a full 24-hour and 7 days period for each group
    result_series = grouped_data.apply(lambda group: (
        group['time_diff'].sum() >= 24 * 60 * 60 * 7) and
        (group['startTime'].min().hour == 0) and
        (group['endTime'].max().hour == 23)
    )

    return result_series

# Adjust the column names based on your specific dataset structure
df = pd.read_csv(r"C:\Users\amrit\Downloads\MapUp-Data-Assessment-F-main\MapUp-Data-Assessment-F-main\datasets\dataset-2.csv")
result_time_check = time_check(df)

# Print the result time check series
print(result_time_check)
```
